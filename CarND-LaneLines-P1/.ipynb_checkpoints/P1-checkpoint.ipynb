## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, regionOfInterest, color=[255, 0, 0], thickness=4):
    """Draws two lines on the left and right of the lane based 
    on the line segments identified with the Hough Transform.
    """
    leftlane = []
    rightlane = []
    
    """This loop checks if endpoints of line segments are in the 'region of interest',
    and if so calculates their slope and intercept. Based on the slope being
    either positive or negative, it will add the slope m and intercept b pairs to the
    left or right lane arrays
    """
    for line in lines:
        for x1, y1, x2, y2 in line:
            if cv2.pointPolygonTest(regionOfInterest, (x1,y1), False)>0:
                if cv2.pointPolygonTest(regionOfInterest, (x2,y2), False)>0:
                    m = ((y2-y1)/ (x2-x1))
                    b = (y1 - (m*x1))
                    if m < 0:
                        leftlane.append([m, b])
                    else:
                        rightlane.append([m, b])
    
    """The values below will ultiately define the endpoints on the final left and right lane 
    lines. The vertical posiiton of the lines are fixed, from the top of the region of 
    interest to the bottom of the image. The horizontal coordinates are calculated
    in the next steps.
    """
    top_ROI = 0.6*imshape[0]
    bottom_ROI = imshape[0]
    
    """The average of left and right lane line slopes and interceps are calculated. 
    The horizontal endpoints are calculated using the average m and b values, and the 
    final lane lines are drawn.
    """
    l_len = len(leftlane)
    if (l_len > 0):
        l_sum = [sum(x) for x in zip(*leftlane)]
        avg_left_m = l_sum[0]/l_len
        avg_left_b = l_sum[1]/l_len
        left_x1 = ((top_ROI - avg_left_b)/ avg_left_m)
        left_x2 = ((bottom_ROI - avg_left_b)/ avg_left_m)
        cv2.line(img, (int(left_x1), int(top_ROI)),  (int(left_x2), bottom_ROI),  [255, 0, 0], 2)
    
    r_len = len(rightlane)
    if (r_len > 0):   
        r_sum = [sum(x) for x in zip(*rightlane)]
        avg_right_m = r_sum[0]/r_len
        avg_right_b = r_sum[1]/r_len 
        right_x1 = ((top_ROI - avg_right_b)/ avg_right_m)
        right_x2 = ((bottom_ROI - avg_right_b)/ avg_right_m)
        cv2.line(img, (int(right_x1), int(top_ROI)), (int(right_x2), bottom_ROI), [255, 0, 0], 2)
        

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, regionOfInterest):
    """
    NOTE: Function has been modified to only find lines in the 
          region of interest, defined by 'regionOfInterest'. 
    `img` should be the output of a Canny transform.   
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines, regionOfInterest)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
````

````python
def process_image(image):
    # Convert to grayscale
    image_gray = grayscale(image)
    
    # Apply Gaussian Blur
    kernel_size = 7
    image_blur = gaussian_blur(image_gray, kernel_size)
    
    # Apply Canny Edge Detection
    image_edges = canny(image_blur, 50, 120)
    
    # Define coordinates of our Region of Interest (ROI)
    imshape = image.shape
    vertices = np.array([[(.45*imshape[1], 0.6*imshape[0]),
                          (0.1*imshape[1], 1*imshape[0]), 
                          (1*imshape[1], 1*imshape[0]), 
                          (0.55*imshape[1], 0.6*imshape[0])]], dtype=np.int32)
    
    # Define Hough Transform parameters
    rho = 1
    theta = np.pi/180
    threshold = 30
    min_line_len = 30
    max_line_gap = 1
    
    # Acquire an array of hough line endpoints
    lines = cv2.HoughLinesP(image_edges, rho, theta, threshold, np.array([]), min_line_len, max_line_gap)
    
    # Draw lane lines on the frame
    draw_lines(image, lines, vertices, [255, 0, 0], 4)
    
    # Return original image with lines drawn
    return image
````
